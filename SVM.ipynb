```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
# Target variables will be 0,1,2
```

```python
iris.target_names
```

So 0 will be 'setosa', 1 will be versicolor, 2 will be virginica

```python
df[df.target==1].head()
```

```python
df[df.target==2].head()
```

So in the given data 0 to 49 will be Setosa
50 to 99 will be Versicolor and
100 to 150 will be Virginica

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
X = df.drop(['target','flower_name'],axis = 'columns')
X.head()
```

```python
y = df.target
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
len(X_train),len(X_test)
```

```python
from sklearn.svm import SVC
model=SVC()
```

```python
model.fit(X_train,y_train)
```
